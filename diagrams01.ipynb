```haskell
import Diagrams.Prelude
import IHaskell.Display
import IHaskell.Display.Diagrams
import Diagrams.Backend.Cairo

-- Add attributes to the drawing on the diagram

example1 :: Diagram B R2
example1 = circle 1 # fc blue
                    # lw 0.05
                    # lc purple
                    # dashing [0.2,0.05] 0

diagram example1
```

```haskell
import Diagrams.Prelude
import IHaskell.Display
import IHaskell.Display.Diagrams
import Diagrams.Backend.Cairo

-- 
example2 :: Diagram B R2
example2 = circle 1 # fc red # lw 0 ||| circle 1 # fc green # lw 0

diagram example2
```

```haskell
-- We can draw diagrams, right in the notebook.
:extension NoMonomorphismRestriction TypeSynonymInstances FlexibleInstances

-- import Diagrams.Prelude
-- import IHaskell.Display
import IHaskell.Display.Diagrams
import Diagrams.Backend.Cairo

-- By Brent Yorgey
-- Draw a Sierpinski triangle!

sierpinski 1 = eqTriangle 1
sierpinski n =     s
                  ===
               (s ||| s) # centerX
  where s = sierpinski (n-1)
  
  
instance Show (Diagram Cairo R2) where
  show x = "I'am a diagram"

-- The `diagram` function is used to display them in the notebook.
diagram $ sierpinski 4
            # centerXY
            # fc black
          `atop` square 10
                   # fc white
                   
```
